```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
plt.plot(x,y)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('My Awesome Line')
plt.show()
```

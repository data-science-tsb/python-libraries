# Matplotlib
Click [ME](https://matplotlib.org/gallery/index.html) for more examples

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
plt.plot(x,y)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('My Awesome Line')
plt.show()
```

#### Subplot
[Multiple](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html) plots in one line

```python
plt.subplot(1,2,1) #row,cols,plot index
plt.plot(x,y, 'r')

plt.subplot(1,2,2) #row,cols,plot index
plt.plot(x,y, 'b')
plt.show()
```

#### Using OOP
The proper way of using Matplotlib

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height
axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height
axes1.plot(x,y)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y,x)
```
